```python
# Define parabolic curve in 2D
def Linear(x,a,b):
    y = a+b*x
    return y
# Define parabolic curve in 3D
def Parabola(x,a,b,c):
    # b: speed
    # 2c : acceleration
    y = a+b*x+c*x**2
    return y
```

```python
#Check if X and T are parabolic.
paramsTX, pcovTX = curve_fit(Parabola, dataT, dataX)
dataTX_fit = Parabola(dataT, paramsTX[0], paramsTX[1],paramsTX[2])
plt.figure()
plt.plot(dataT,dataX)
plt.plot(dataT,dataTX_fit)
plt.show()
print(paramsTX)
# It seems there was a "twist" around T = 1250 or X = 0.018
```

```python
tmp_A = []
tmp_b = []
for i in range(len(dataX1)):
    tmp_A.append([dataX1[i], dataY1[i], 1])
    tmp_b.append(dataZ1[i])
b = np.matrix(tmp_b).T
A = np.matrix(tmp_A)
fit = (A.T * A).I * A.T * b
errors = b - A * fit
residual = np.linalg.norm(errors)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fx = fit[0].item(0)
fy = fit[1].item(0)
fc = fit[2].item(0)
print "solution:"
print "%f x + %f y + %f = z" % (fit[0], fit[1], fit[2])
print "errors:"
print errors
print "residual:"
print residual
xx, yy = np.meshgrid(np.arange(-1,2), np.arange(-1,1))
zz = fx*xx+fy*yy+fc
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(xx,yy,zz, color='blue')
plt.show()
```

```python
plt.figure()
ax = plt.subplot(111, projection='3d')
ax.scatter(dataX1, dataY1, dataZ1,color='b')
ax.plot_surface(xx, yy, zz, color='cyan')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(30, 30)
plt.show()
```
