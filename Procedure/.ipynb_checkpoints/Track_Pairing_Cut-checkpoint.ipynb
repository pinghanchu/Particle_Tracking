```python
df = pd.read_csv('TrackPair_PID.csv')
tL = pd.read_csv('trackL_inv_clean.csv')
tR = pd.read_csv('trackR3_frameless.csv')
tL.head()
```

```python
Track = df[(df['FF Ave']<5)&(df['Mass Diff']<50)&(df['Frame Count']>1000)&(df['FF Small Count']>100)]
```

```python
Track
```

```python
Pair = []
for iL in Track['PIDL'].unique():
    PL = Track[Track['PIDL']==iL]
    TrackL = tL[tL['particle']==iL]
    for iR in PL['PIDR'].unique():
        TrackR = tR[tR['particle']==iR]
        for iframeL in TrackL['frame'].unique():
            TrackL1 = TrackL[TrackL['frame']==iframeL]
            ixL = TrackL1['x'].iloc[0]
            iyL = TrackL1['y'].iloc[0]  
            imassL = TrackL1['mass'].iloc[0]
            TrackR1 = TrackR[TrackR['frame']==iframeL]             
            if(len(TrackR1)>0):
                ixR = TrackR1['x'].iloc[0]
                iyR = TrackR1['y'].iloc[0] 
                imassR = TrackR1['mass'].iloc[0]
                ipL = [ixL,iyL,1]                       
                ipR = [ixR,iyR,1]
                LineL = np.dot(ipR,MatrixL) 
                LineR = np.dot(ipL,MatrixR) 
                dlr = math.fabs(np.dot(ipL,LineL))/math.sqrt(LineL[0]*LineL[0]+LineL[1]*LineL[1])
                drl = math.fabs(np.dot(ipR,LineR))/math.sqrt(LineR[0]*LineR[0]+LineR[1]*LineR[1])
                FF = dlr + drl
                ipair = [iL,ixL,iyL,imassL,dlr,iR,ixR,iyR,imassR,drl]
                Pair.append(ipair)
                P1 = pd.DataFrame(Pair)
                P1.columns = ['PIDL','XL','YL','MassL','DistanceL','PIDR','XR','YR','MassR','DistanceR']
                P1.to_csv("Pair.csv")
```
